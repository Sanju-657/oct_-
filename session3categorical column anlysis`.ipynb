```python
l1=[]
for i in cat_cols[1:]:
    labels=visa_df[i].unique()
    print(f"{labels} are unique elements of {i}column")
    for j in labels:
        cond=visa_df[i]==j
        count=len(visa_df[cond])
        print(f"{count} applications info is {j}")
        l1.append(count)
        pd.DataFrame(zip(labels,l1),columns=[f'{i}','count'])
```

```python
l=[]
labels=visa_df['continent'].unique()
for i in labels:
    con=visa_df['continent']==i
    count=len(visa_df[con])
    l.append(count)
pd.DataFrame(zip(labels,l),columns=['continent','count'])
```
