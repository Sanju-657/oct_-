# Issues in **Above** Code
# Appending outside the loop:

# l1.append(count) is not inside the inner loop, so it only stores the last count for each column.
# Incorrect DataFrame creation:

# zip(labels, l1) causes issues because l1 keeps growing, but labels is fixed.
# Inefficient Filtering (con= churn_df[i]==j)

# Instead of churn_df[con], directly count sum(con), which is much faster.

```python
l1 = []  # List to store counts

for i in cat_cols:
    labels = churn_df[i].unique()  # Get unique categories for the column
    print(f"{labels} are the unique elements in the {i} column")
    
    temp_list = []  # Temporary list for each column
    
    for j in labels:
        count = sum(churn_df[i] == j)  # Count occurrences efficiently
        temp_list.append((j, count))  # Store as tuple
    
    l1.extend(temp_list)  # Add to main list
    # Creating DataFrame
    df = pd.DataFrame(temp_list, columns=[f'{i}', 'count'])
    print(df)
print(l1)  # List of all (category, count) pairs across all categorical columns
```
